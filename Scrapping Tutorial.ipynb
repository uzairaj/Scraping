PRINT SINGLE PARAGRAPH DATA IN A DIV

```python
data = soup.find('div',attrs={"class":"mw-parser-output"})
print(data.p.text)
```

PRINT ALL NESTED PARAGRAPH DATA 

```python
data = soup.find_all('div',attrs={"class":"mw-content-ltr"})

for s in data:
    chk=s.find_all('p')
    for i in chk:
        print(i.text)
```

PRINT SINGLE HEADING IN A DIV

```python
data = soup.find('span',attrs={"class":"mw-headline"})
print(data.text)
```

PRINT ALL NESTED HEADING

```python
data = soup.find_all('div',attrs={"class":"mw-content-ltr"})

for s in data:
    chk = s.find_all('span',attrs={"class":"mw-headline"})
    for i in chk:
        print(i.text)
```

```python
website_url = requests.get('https://osd.dms.myflorida.com/directories/results?directory%5Bcommodity_code%5D=&directory%5Bcommodity_description%5D=&directory%5Bcounty%5D=&directory%5Bdesignation%5D=&directory%5Bvendor_name%5D=')
soup = BeautifulSoup(website_url.content, 'html.parser')
print(soup.prettify())
```

```python
data = []
tb = soup.find('table', class_='table table-bordered table-striped dataTable')
table_body = tb.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])
```

```python
scrap_data = pd.DataFrame(data, columns = ['Name','Contact','Adress','City','State','Phone'])
scrap_data.to_excel('scraped_data.xlsx',index=False)
```
